# Datamining bestaande dataset - Million Song Dataset (MSD) + Spotify features

### Dataset inladen

```python
import pandas as pd

df = pd.read_csv("msd_spotify_features.csv")
df.head()
```

### Mini-dataset aanmaken om code op te testen

```python
df_small = df.sample(5000, random_state=42)
df_small.to_csv("msd_small_5000.csv", index=False)
```

### Filter functies voor de dataset

```python
def filter_relaxing(df):
    return df[
        (df["tempo"].between(50, 70)) &
        (df["energy"] < 0.4) &
        (df["acousticness"] > 0.5)
    ]

def filter_energizing(df):
    return df[
        (df["tempo"].between(120, 150)) &
        (df["energy"] > 0.6)
    ]


```

### Gerichte zoekopdrachten naar nummers met bepaalde kenmerken

```python
def search_by_features(df, min_tempo=None, max_tempo=None, min_energy=None, max_energy=None):
    result = df.copy()
    if min_tempo is not None:
        result = result[result["tempo"] >= min_tempo]
    if max_tempo is not None:
        result = result[result["tempo"] <= max_tempo]
    if min_energy is not None:
        result = result[result["energy"] >= min_energy]
    if max_energy is not None:
        result = result[result["energy"] <= max_energy]
    return result
```

### ISO-playlist generator

```python
def build_iso_playlist(df, phases):
    playlist = pd.DataFrame()
    for phase in phases:
        subset = df[
            df["tempo"].between(phase["min_bpm"], phase["max_bpm"])
        ].sample(phase["count"], replace=False)
        playlist = pd.concat([playlist, subset])
    return playlist
```
