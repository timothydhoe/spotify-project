```python
df = pd.read_csv('../../../../data/check_in/check_in.csv',
                 parse_dates=['Tijdstempel'],
                 dayfirst=True)
```

```python
# Vandaag definiëren
vandaag = pd.Timestamp.today().normalize()


# Laatste luistersessie per deelnemer
laatste = (
    df.groupby('Deelnemerscode')['Tijdstempel']
      .max()
      .reset_index()
)

print(laatste)
```

```python
laatste['dagen_sinds'] = (vandaag - laatste['Tijdstempel']).dt.days.clip(lower=0)
laatste
```
