```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
calm_df = pd.read_csv("/Users/timothydhoe/Code/spotify-project/docs/td_playground/TEST-001_calm_playlist.csv")
upbeat_df = pd.read_csv("/Users/timothydhoe/Code/spotify-project/docs/td_playground/TEST-001_upbeat_playlist.csv")
```

```python
# Print the first 10 songs of the calm playlist
print(calm_df.head(10))
```

```python
# Print the first 10 songs of the upbeat playlist
print(upbeat_df.head(10)) 
```

```python
#
calm_df["playlist"] = "Calm"
upbeat_df["playlist"] = "Upbeat"

df = pd.concat([calm_df, upbeat_df], ignore_index=True)
```

```python
# Print out visual of the Feature differences
plt.figure()
for label, group in df.groupby("playlist"):
    plt.hist(group["valence"], alpha=0.5, label=label)

plt.xlabel("Valence")
plt.ylabel("Count")
plt.title("Valence Distribution by Playlist")
plt.legend()
plt.show()
```

```python
# Print out visual of the Feature differences
plt.figure()
for label, group in df.groupby("playlist"):
    plt.hist(group["tempo"], alpha=0.5, label=label)

plt.xlabel("Tempo")
plt.ylabel("Count")
plt.title("Tempo Distribution by Playlist")
plt.legend()
plt.show()
```
